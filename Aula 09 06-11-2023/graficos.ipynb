###Gr√°fico simples

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# criando array aleat√≥rio para gerar os gr√°ficos

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Criando os gr√°ficos com os valores acima

plt.plot(x, y)
```

```python
plt.plot(x, y, marker = '*')
```

```python
plt.plot(
    x,
    y,
    marker = '*', # Estilo dos pontos
    c = 'blue', # Cor da linha
    ds = 'steps', # Forma de liga√ß√£o
    ls = '--' # Estil oda linha
)
```

```python
plt.plot(
    x,
    y,
    marker = '*', # Estilo dos pontos
    c = 'red', # Cor da linha
    ds = 'steps', # Forma de liga√ß√£o
    ls = '--', # Estilo da linha
    mec = 'blue' # Cor dos marcadores
)
```
